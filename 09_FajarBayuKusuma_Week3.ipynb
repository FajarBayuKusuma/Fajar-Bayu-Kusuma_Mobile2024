```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow #forImage display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

**Transformasi Linier Brightness**

Formula: *g(x,y)*=*f(x,y)+b*
*g(x,y)* adalah nilai pixel setelah tranformasi, *f(x,y)* adalah nilai pixel asli, sedangkan *b* adalah nilai brightness

```python
print(' Mengubah Tingkat Kecerahan Citra  ')
print('-----------------------------------')
try:
  brightness = int(input('Masukkan Nilai Kecerahan: '))
except ValueError:
  print('Error, not a numberr')

original = cv.imread('/content/drive/MyDrive/KULIAH/Semester 5/PCVK/Week 3/female.tiff')
brightness_image = np.zeros(original.shape, original.dtype)

#Akses oer pixel
for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y,x,c] = np.clip(original[y,x,c] + brightness, 0, 255)

#Fungsi np.clip() memastikan nilai tersebut tetap berada dalam
#rentang 0 hingga 255 (karena nilai piksel gambar berkisar antara 0 hingga 255).
#cara simpel tanpa for loop
#brigthness_image = cv.convertScaleAbs(original, beta=brightness)

final_frame = cv.hconcat((original, brightness_image))
cv2_imshow(final_frame)
```

4. Implementasikan transformasi grayscale menggunakan metode averaging, lightness, dan luminance pada Google Colaboratory menggunakan formula yang terdapat pada bagian Ulasan Teori, sehingga menghasilkan keluaran seperti pada laman tugas.

```python
# A. averaging
averaging = cv.imread('/content/drive/MyDrive/KULIAH/Semester 5/PCVK/Week 3/KTP3.jpg')

average_image = np.zeros(averaging.shape[:2], dtype=np.uint8)

for y in range(averaging.shape[0]):
    for x in range(averaging.shape[1]):
        red = int(averaging[y, x, 2])
        green = int(averaging[y, x, 1])
        blue = int(averaging[y, x, 0])
        avg = (red + green + blue) // 3
        average_image[y, x] = np.uint8(avg)


plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv.cvtColor(averaging, cv.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(average_image, cmap='gray')
plt.title('Averaging')


plt.show()
```
